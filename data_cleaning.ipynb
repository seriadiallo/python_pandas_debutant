####  pour regler le probleme d'execo dont les cellules ce sont deplacees
####  exemple index 6, pas de rank et de point et les valeur de player et team se sont deplacees
#### ce sont des execo donc, rank et points sont egaux aux valeurs directement precedentes

```python

f = df.loc[0]
for index in range(df.shape[0]):
    if np.where(df.loc[index].isna())[0].shape[0] != 0:
        df.loc[index, 'team'] = df.loc[index, 'player']
        df.loc[index, 'player'] = df.loc[index, 'rank']
        df.loc[index, 'rank'] = f['rank']
        df.loc[index, 'points'] = f['points']
    f = df.loc[index]
df.head(10)
```

#### les deux colonnes pourcentages et pourcentage ne sont pas conformes
nous allons renomer 'pourcentage' en 'percentages' et supprimer 'pourcentages'

```python
# cette cellule ne doit etre executee qu'une seule fois
df.rename(columns={'pourcentage': 'percentages'}, inplace=True)
df.drop(columns=['pourcentages'], inplace=True)
```

suppremer les valeurs en pourcentage dans la colonne points

```python
df['points'].where(df['points'].str.find('%') == -1, np.nan, inplace=True)
df.loc[170:183]
```

```python
df['points'] = df['points'].astype('int64', errors='ignore')
```

```python
df['points'] = pd.to_numeric(df['points'], downcast='integer', errors='ignore')
df.info()
```

```python
grouped = df.groupby('year')
sum_grouped = grouped.apply(lambda x: x.sum(numeric_only=True))
sum_grouped.head()
```

```python
sum_grouped.info()
```

```python
for index in range(df.shape[0]):
    if pd.isna(df.loc[index, 'points']) is False:
#        calculer le pourcentage et garder selement deux chiffres apres la virgule 
        df.loc[index, 'percentages'] = round(df.loc[index, 'points'] * 100 / sum_grouped.loc[df.loc[index, 'year'], 'points'], 2)
        df.loc[index, 'points'] = int(df.loc[index, 'points']) # garder seulement la partie entiere
df.sample(n=10)
```

#### Supprimer le '%' dans la colonne percentages

```python
cleanig_percentages = lambda elem: str(elem)[:str(elem).find('%')] if str(elem).endswith('%') else elem
df['percentages'] = df['percentages'].apply(cleanig_percentages)
df.sample(n=8)
```

### certains valeur de player contien des [ not..] 

nous allons les supprimer

```python
# s'il contient, retourner les carcteres avant les corchets, sinon retourner toute la chaine
df['player'] = df['player'].apply(lambda elem: elem[:elem.find('[')] if elem.find('[') >= 0 else elem)
df.sample(n=12)
```
